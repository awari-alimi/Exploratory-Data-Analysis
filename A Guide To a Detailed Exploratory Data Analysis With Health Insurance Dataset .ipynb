# Exploratory Data Analysis (EDA)

## What is EDA?
Exploratory Data Analysis (EDA) is employed to determine the most effective ways to manipulate data sources to acquire valuable insights. EDA involves investigating data sets to discover patterns, spot anomalies, test hypotheses, and check assumptions. This analysis enhances the understanding of data set variables and their relationships, often presented through summaries, tables, and visualizations.

## Types of exploratory data analysis
There are four primary types of EDA:

- Univariate non-graphical. This is simplest form of data analysis, where the data being analyzed consists of just one variable to describe the data and find patterns that exist within it.

- Univariate graphical. Since Non-graphical methods don’t provide a full picture of the data, graphical methods are therefore required. Common types of univariate graphics include: Stem-and-leaf plots, Histograms, a bar plot and Box plots

- Multivariate nongraphical: Multivariate data arises from more than one variable. Multivariate non-graphical EDA techniques generally show the relationship between two or more variables of the data through cross-tabulation or statistics.

- Multivariate graphical: Multivariate data uses graphics to display relationships between two or more sets of data. The most used graphic is a grouped bar plot or bar chart with each group representing one level of one of the variables and each bar within a group representing the levels of the other variable. Other common type of multivariate graphics is Scatter plot, Multivariate chart, Run chart, Bubble chart and Heat map.

```python
# Add labels and title
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Advanced Scatter Plot with Regression Line for Charges vs. Age')
```

```python
# Show plot
plt.show()
```

```python
Explanation of Enhancements:
•	Hue (hue='smoker'): Colors the points based on whether the individual is a smoker, helping differentiate groups.
•	Style (style='sex'): Changes the marker style based on gender, adding another dimension of information to the plot.
•	Size (size='bmi'): Scales the marker size based on the bmi column, allowing you to visualize another variable dimension.
•	Palette (palette='viridis'): Changes the color palette to 'viridis' for better color contrast.
•	Markers (markers=['o', 's']): Specifies different marker shapes for male and female.
•	Regression Line (sns.regplot(...)): Adds a regression line to show the trend, with a customized dashed line style.
•	Alpha (alpha=0.7): Sets transparency to handle overlapping data points.
```
